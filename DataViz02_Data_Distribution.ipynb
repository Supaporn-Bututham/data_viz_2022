#Quiz กลุ่ม final

1. เขียน function ที่รับ input เป็น output ของ boxplot แล้ว แสดงค่า min,max,q1,q2,q3
2. วาด boxplot เปรียบเทียบการกระจายตัวของข้อมูล 
PetalLength ของดอก iris ทั้ง 3 ชนิด

## ข้อ 1

```python
def box_vals(ybp) :  # สร้าง function box_vals ขึ้นมา โดยรับ input จาก ybp
    q1 = max(ybp['whiskers'][0].get_xdata()) #ค่าที่มากสุดของหนวดแมวทางซ้าย
    q2 = ybp['medians'][0].get_xdata()[0] # ค่ากลางของข้อมูล
    q3 = min(ybp['whiskers'][1].get_xdata()) # ค่าที่น้อยที่สุดของหนวดแมวด้านขวา
    mmin = min(ybp['whiskers'][0].get_xdata()) #  ค่าที่น้อยที่สุดของหนวดแมวด้ายซ้าย
    mmax = max(ybp['whiskers'][1].get_xdata()) # ค่าที่มากที่สุดของหนวดแมวด้านขวา
    print(f' Q1 = {q1} \n Q2 = {q2} \n Q3 = {q3} \n min = {mmin} \n max = {mmax}')
```

```python
box_vals(ybp) 
```

```python
plt.figure(figsize =(15, 5)) # กว้าง15 และ สูง 5

plt.subplot(1, 3, 1)  # กำหนดตำแหน่งกราฟที่ 1  คือ  1 แถว 3 คอลัมน์  กราฟนี้อยู่ตำแหน่งที่ 1
plt.boxplot(df['PetalLength'][0:50],vert=False);  #ชนิด 1
plt.title('setosa') 
 
plt.subplot(1, 3, 2)  # กำหนดตำแหน่งกราฟที่ 2  คือ  1 แถว 3 คอลัมน์  กราฟนี้อยู่ตำแหน่งที่ 2
plt.boxplot(df['PetalLength'][50:100],vert=False); # ชนิด 2
plt.title('versicolor') 

plt.subplot(1, 3, 3) # กำหนดตำแหน่งกราฟที่ 3  คือ  1 แถว 3 คอลัมน์  กราฟนี้อยู่ตำแหน่งที่ 3
plt.boxplot(df['PetalLength'][100:150],vert=False); # ชนิด 3
plt.title('virginica');
```

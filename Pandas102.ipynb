## ต่อแกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head()
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

## Dictionary(row oriented) วิธีที่ 1

```python
data_list = [ {'account':'Jones LLC','Jan':150,'Feb':200,'Mar':140}, # สร้าง list ขึ้นมาหนึ่งตัว
                   {'account':'Alpha Co','Jan':200,'Feb':210,'Mar':215},
                   {'account':'Blue Inc','Jan':50,'Feb':90,'Mar':95},
                ]
```

```python
data_list
```

```python
df1 = pd.DataFrame(data_list)
df1
```

## List(row oriented) วิธีที่ 2

```python
data = [['Jones LLC',150,200,140],
            ['Alpha Co',200,210,215],
            ['Blue Inc',50,90,95]] # แต่ละ record
header = ['account','Jan','Feb','Mar'] # หัวคอลัมน์
```

```python
df2 = pd.DataFrame.from_records(data,columns=header) # กำหนดให้คอลัมน์คือตัวแปร header
df2
```

## Dictionary(column oriented) วิธีที่ 3

```python
data_dict = {'account':['Jones LLC','Alpha Co','Blue Inc'], # สร้างแต่ละคอลัมน์
                    'Jan':[150,200,50],
                    'Feb':[200,210,90],
                    'Mar':[140,215,95]
}
```

```python
df3 = pd.DataFrame.from_dict(data_dict)
df3
```

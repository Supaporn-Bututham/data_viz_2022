# **โหลดข้อมูล Covid-19 ทั้ง 4 ชุด**

### **คำอธิบาย คอลัมน์**

* No : รหัสผู้ป่วย
* Announce Date : วันที่แถลงข่าว
* Notification Date : วันที่กระทรวงได้รับรายงาน
* Sex : เพศผู้ป่วย
* Age : อายุของผู้ป่วย
* Nationality : สัญชาติ
* Province of isolation : จังหวัดที่รักษาตัว
* Province of onset จังหวัดที่เริ่มพบว่าตัวเองมีอาการป่วย
* District of onset อำเภอที่เริ่มพบว่าตัวเองมีอาการป่วย

## **ชุดข้อมูลที่ 1**

```python
data_01 = pd.read_csv(os.path.join(path,'confirmed-cases.csv')) # ไฟล์ที่เราต้องการ
data_01 # print data ดูว่ามันหน้าตาเป็นยังไง
```

## **ชุดข้อมูลที่ 2**

```python
data_02 = pd.read_csv(os.path.join(path,'confirmed-cases-since-120864.csv')) # ไฟล์ที่เราต้องการ
data_02 # print data ดูว่ามันหน้าตาเป็นยังไง
```

## **ชุดข้อมูลที่ 3**

```python
data_03 = pd.read_csv(os.path.join(path,'confirmed-cases-since-271064.csv')) # ไฟล์ที่เราต้องการ
data_03 # print data ดูว่ามันหน้าตาเป็นยังไง
```

## **ชุดข้อมูลที่ 4**

```python
data_04 = pd.read_csv(os.path.join(path,'confirmed-cases-since-280265.csv')) # ไฟล์ที่เราต้องการ
data_04 # print data ดูว่ามันหน้าตาเป็นยังไง
```

# **รวมข้อมูล Covid-19 ทั้ง 4 ชุดเข้าด้วยกัน**

```python
data_covid = pd.concat([data_01, data_02, data_03, data_04])
data_covid
```

# **แสดง Columns ที่สนใจ**

```python
data_covid = data_covid[['announce_date','sex','risk','province_of_onset']]
data_covid
```

# **ตรวจสอบค่า Missing ของตาราง data_covid**

```python
data_covid.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง สรุปคือทุกคอลัมน์ไม่มีค่า  missing
```

### **แก้ค่า Missing ของคอลัมน์ 'sex'**

```python
# แทนค่า Missing ของคอลัมน์ 'sex' เป็น 'ไม่ระบุ'
data_covid[['sex']] = data_covid[['sex']].fillna('ไม่ระบุ')
```

```python
data_covid.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'sex' เป็น False นั่นคือไม่มีค่า missing
```

### **แก้ค่า Missing ของคอลัมน์ 'risk'**

```python
set(data_covid['risk'])      # ลองดู set ข้อมูลของคอลัมน์ 'risk'
```

```python
data_covid[['risk']] = data_covid[['risk']].fillna('ไม่มีข้อมูล')  # แทนค่า Missing ของคอลัมน์ 'risk' ด้วย 'ไม่มีข้อมูล'
```

```python
data_covid.isnull().any()       # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'risk' เป็น False นั่นคือไม่มีค่า missing
```
